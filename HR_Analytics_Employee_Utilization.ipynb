```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv(r"C:\\Users\\user\\Manish\\GEMS\\GEMS - Domain Analytics\\HR analytics\\Employee_Utilization.csv")
df.head()
```

### 1. What are the top countries by the number of employees?

```python
top_countries = df['Country'].value_counts()
top_countries
```

## 2. Is there a correlation between salary and average utilization 

```python
correlation = df['Salary'].corr(df['Average Utilisation'])
correlation
```

## 3. Which souring method is most effective in hiring?

```python
sourcing_effectiveness = df['Sourced through'].value_counts()
print(sourcing_effectiveness)
# visualising the sourcing effectiveness 
sourcing_effectiveness.plot(kind='bar',xlabel='Sourcing Method',ylabel ='Frequency',title ='Sourcing Method Effectiveness')
plt.show()
```

## 4. What is the distribution of employees previous year rating?

```python
df['Previous Year Rating'].hist(bins = 5)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Previous Year Ratings')
plt.show()
```

## 5. What proportion of employees were recently promoted?

```python
promotion_proportion = df['Promoted Recently'].mean() * 100 # calculating as a percentage 
promotion_proportion
```

## 6. How does the number of projects relate to salary?

```python
plt.scatter(df['No of Projects'],df['Salary'])
plt.xlabel('Number of Projects')
plt.ylabel('Salary')
plt.title('Relationship between number of projects and Salary')
plt.show()
```

```python
correlation = df['Salary'].corr(df['No of Projects'])
correlation
```

## 7. To check whether there is any difference in average utilization between different working modes?

```python
utilization_by_mode = df.groupby('Working Mode')['Average Utilisation'].mean()
print(utilization_by_mode)
# visualizing the average utilisation by working mode 
utilization_by_mode.plot(kind = 'bar',xlabel = 'Working Mode',ylabel = 'Average Utilization',title = 'Avg utilization by working mode')
plt.show()
```

## 8. What is the distribution of salaries across different base locations?

```python
plt.figure(figsize=(20,10))
sns.boxplot(data=df,x='Base Location',y = 'Salary')
plt.xticks(rotation=45)
plt.xlabel('Base Location')
plt.ylabel('Salary')
plt.title('Salary distribution across base locations')
plt.show()
```

## 9. Do Employees who are sourced through referal have a higher salary on average?
